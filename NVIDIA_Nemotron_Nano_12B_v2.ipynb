```
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="nvidia/NVIDIA-Nemotron-Nano-12B-v2")
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
```
